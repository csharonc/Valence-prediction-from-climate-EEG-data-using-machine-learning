##### Make a trigger dictionary: stimuli_dict
- Dict names are based on the names found in SourceStimuliName
- They're all given a unique value, but not yet grouped
- Key = str, value = int (correct)

#### Stimuli dict has a logic to it    
The first number indicates the type of stimulus   

1 = Anything we don't use  
2 = baseline from the beginning (before any stimuli are shown)  
     - second number 1 = baseline with eyes open  
     - second number 2 = baseline with eyes closed  
3 = grijsscherm  
4 = fixcross  
5 = positive image  
6 = neutral image  
7 = negative image  

The last 3 numbers are simply the index number of the stimulus name

```python
stim_array = res1["SourceStimuliName"].unique()

stimuli_triggers = pd.DataFrame({"StimuliNames":stim_array})
stimuli_triggers["Value"] = range(10001, 10001 + len(stimuli_triggers))
stimuli_triggers[""] = 0
stimuli_triggers[" "] = 0

stimuli_dict = dict(zip(stimuli_triggers["StimuliNames"], stimuli_triggers["Value"]))

for key, val in stimuli_dict.items():
    if "baseline_" in key:
        print(f"Processing key: {key}, value: {val}")
        value_str = str(val)
        new = "2" + value_str[1:]
        stimuli_dict[key] = int(new)
        if "baseline_open" in key:
            value_str = str(val)
            new2 = new[0] + "1" + new[2:]
            stimuli_dict[key] = int(new2)
        if "baseline_closed" in key:
            value_str = str(val)
            new2 = new[0] + "2" + new[2:]
            stimuli_dict[key] = int(new2)
    if key.startswith("grijsscherm"):
        value_str = str(val)
        new = value_str.replace(value_str[0], "3")
        stimuli_dict[key] = int(new)
    if key.startswith("fixcross"):
        value_str = str(val)
        new = value_str.replace(value_str[0], "4")
        stimuli_dict[key] = int(new)
    if key.startswith("Pos"):
        value_str = str(val)
        new = value_str.replace(value_str[0], "5")
        stimuli_dict[key] = int(new)
    if key.startswith("Neutral"):
        value_str = str(val)
        new = value_str.replace(value_str[0], "6")
        stimuli_dict[key] = int(new)
    if key.startswith("Neg"):
        value_str = str(val)
        new = value_str.replace(value_str[0], "7")
        stimuli_dict[key] = int(new)


stimuli_dict
for key, value in stimuli_dict.items():
    print(key, value)
```
